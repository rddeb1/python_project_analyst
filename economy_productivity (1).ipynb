```python
city = df.groupby(['ï»¿City'], as_index=False)['R&D Expenditure (% of GDP)'].sum().sort_values(by='R&D Expenditure (% of GDP)', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='ï»¿City', y='R&D Expenditure (% of GDP)', hue='ï»¿City', data=city)
```

R&D Expenditure (% of GDP) as per the state total rate%

```python
year = df.groupby(['Year'], as_index=False)['R&D Expenditure (% of GDP)'].mean()

sns.set(rc={'figure.figsize':(15,6)})
sns.barplot(x='Year', y='R&D Expenditure (% of GDP)',hue='Year', data= year)
```

R&D Expenditure (% of GDP) over the year 

```python
Unemployment = df.groupby(['ï»¿City', 'Year'], as_index=False)['Unemployment Rate (%)'].max().head(100)
sns.set(rc={'figure.figsize':(30,5)})
sns.barplot(data=Unemployment,x='ï»¿City', y='Unemployment Rate (%)', hue='ï»¿City')
```

In the above graph we have showed the unemployement rate as per the state 

```python
Unemployment = df.groupby([ 'Year'], as_index=False)['Unemployment Rate (%)'].mean().head(100)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=Unemployment,x='Year', y='Unemployment Rate (%)', hue='Year')
```

unemployement rate as per the city by (years)

```python
youth = df.groupby(['ï»¿City'], as_index=False)['Youth Unemployment Rate (%)'].sum()
sns.set(rc={'figure.figsize':(28,5)})
sns.barplot(x='ï»¿City', y='Youth Unemployment Rate (%)',data=youth,hue='ï»¿City')
```

Above graph we have used enmployement rate to know the actual rate of unempy in (CITY)

```python
youth = df.groupby(['ï»¿City'], as_index=False)['Tourism Sector Employment (%)'].sum()
sns.set(rc={'figure.figsize':(30,5)})
sns.barplot(x='ï»¿City', y='Tourism Sector Employment (%)',data=youth,hue='ï»¿City')
```
