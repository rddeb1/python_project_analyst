```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns 
```

```python
df =pd.read_csv('Diwali Sales Data.csv',encoding='unicode_escape')
```

```python
df.shape
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'], axis=1,inplace=True)
```

```python
df.info()
```

```python
pd.isnull(df)  #any value is there in using isnull
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df["Amount"]=df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.describe()
```

```python
ax = sns.countplot(x='Gender',data= df)
for bars in ax.containers:
    ax.bar_label(bars)
    
```

```python
sales_gen = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Gender', y= 'Amount',data= sales_gen)
```

From above graps we can see that most of the buyers are females and even the purchasing power of females are greater then man

```python
df.columns
```

```python
ax = sns.countplot(data=df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group', y= 'Amount',data= sales_gen)
```

Most of the amt spend as per age

```python
sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(19,5)})
sns.barplot(data = sales_state, x='State', y= 'Orders', hue='State')
```

its a state based on highest orders 

```python
df.columns
```

```python
ax = sns.countplot(data=df, x = 'Marital_Status', hue='Marital_Status')
sns.set(rc={'figure.figsize':(5,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Marid_status = df.groupby(['Marital_Status','Gender'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = Marid_status, x='Marital_Status', y= 'Amount', hue='Gender')
```

```python
ax = sns.countplot(data=df, x = 'Occupation', hue='Occupation')
sns.set(rc={'figure.figsize':(23,8)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.countplot(data=df, x = 'Product_Category', hue='Product_Category')
sns.set(rc={'figure.figsize':(25,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
product =df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by= 'Amount', ascending= False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = product, x='Product_Category', y= 'Amount', hue='Product_Category')
```
